```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/Historical_mortality_rates_of_puerperal_fever#Yearly_mortality_rates_for_birthgiving_women_1833.E2.80.931858_for_first_and_second_clinics'
r  = requests.get(url)
data = r.text
soup = BeautifulSoup(data, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.table.tr.th.text
```

```python
#table_2 = soup.find_all('table')[1]#.tr.th.text
```

```python
#columns = soup.table.tr.find_all('th')#[1].text
```

```python
columns = soup.table.tr.find_all('th')#[1].text

column_names = []

for col in columns:
    column_names.append(col.text)

column_names
```

```python
rows = soup.table.find_all('tr')

table_rows = []
vector = []

for row in rows[1:]:
    data = row.find_all('td')
    vector = []
    for el in data:
        vector.append(el.text)
    table_rows.append(vector)

table_rows
```

```python
Data = pd.DataFrame(table_rows,columns=column_names)
```

```python
Data
```

```python
Data.Year = Data.Month.apply(lambda x: x[-4:]).replace(regex={'\D+': np.nan}).fillna(method='ffill')
```

```python
Data.Month = Data.Month.str.extract('([a-zA-Z]{3})\.*', expand=False)
```

```python
Data = Data.replace('na',0).fillna('')
```

```python
Data
```

```python
Data.to_csv('data/hand_wash.csv')
```
